# K-means algorithm

- 주어진 데이터를 지정된 군집(cluster)개수 `K`로 그룹화
- 알고리즘의 결과는 중심(centroid)라 불리는 K개의 점  
- 데이터들은 K개의 군집 중 하나에만 속할 수 있음
- NP-hardness : 군집 구성시 직접 오차함수를 최소화하려면 계산 비용이 많이 듬
- 휴리스틱(heuristics)한 방법으로 로컬 최소값에 빠르게 수렴하는 기법들이 개발
- 통상적으로 몇 번의 반복만으로 수렴되는 `반복 개선(iterative refinement)`를 많이 사용
  - 초기 단계(0단계) : K개 중심의 초기 집합을 결정
  - 할당 단계(1단계) : 각 데이터를 가장 가까운 군집에 할당
  - 업데이트 단계(2단계) : 각 그룹에 대해 새로운 중심을 계산

K개의 중심을 찾는 방법은 여러 가지가 있으나, 그 중 하나는 임의로 선택  
할당 수정 단계를 알고리즘이 수렴되었다고 간주 될때까지 반복

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

임의의 2000개의 데이터를 생성

```python
num_points = 2000
vectors_set = []

for i in range(num_points):
    if np.random.random() > 0.5:
        vectors_set.append([np.random.normal(0.0, 0.9), np.random.normal(0.0, 0.9)])
    else:
        vectors_set.append([np.random.normal(3.0, 0.5), np.random.normal(1.0, 0.5)])
```

```python
df = pd.DataFrame({"x" : [v[0] for v in vectors_set], "y" : [v[1] for v in vectors_set] })
sns.lmplot("x", "y", data=df, fit_reg=False, size=6)
plt.show()
```

4개의 군집으로 그룹화하는 `K-means algorithm`

```python
vectors = tf.constant(vectors_set)

print('sample vector : ',vectors.get_shape())

k = 4
centroides = tf.Variable(tf.slice(tf.random_shuffle(vectors),[0,0],[k,-1]))

print('centroides : ',centroides.get_shape())

expanded_vectors = tf.expand_dims(vectors, 0)
expended_centroides = tf.expand_dims(centroides, 1)

print('expanded_vectors : ',expanded_vectors.get_shape())
print('expended_centroides : ',expended_centroides.get_shape())


assignments = tf.argmin(
    tf.reduce_sum(
        tf.square(
            tf.sub(
                expanded_vectors, expended_centroides)),
        2),
    0)

means = tf.concat(0,
                [tf.reduce_mean(
                    tf.gather(
                        vectors,
                        tf.reshape(
                            tf.where(
                                tf.equal(assignments, c)), [1,-1])),
                    reduction_indices=[1])
                for c in range(k)])

update_centroides = tf.assign(centroides, means)

init_op = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init_op)

for step in range(100):
    _, centroides_values, assignment_values = sess.run([update_centroides, centroides, assignments])
```

- `vectors`와 `centroids`의 차원이 달라서 뺄샘을 할수 없으므로 둘 다 2차원에서 3차원으로 차원을 확장
  - `D2` 차원은 같지만, 확장된 `D0`, `D1` 차원에서는 크기를 결정할 수 없는 차원이 있으므로 상대편 차원만큼 늘려서 계산을 함 

-  `유클리드 제곱거리(squared Euclidean distance)`
  - 1단계 : `tf.sub` 뺄샘
  - 2단계 : `tf.square` 제곱
  - 3단계 : `tf.reduce_sum` 지정한 차원에 따라 원소를 더함
  - 4단계 : `tf.argmin` 지정한 차원을 따라 가장 작은 값의 원소가 있는 인덱스를 리턴

$${ d }^{ 2 }(vector, centroid) = { ({ vector }_{ x }-{ centroid }_{ x }) }^{ 2 }+{ ({ vector }_{ y }-{ centroid }_{ y }) }^{ 2 }$$

- 새로운 중심 계산
  - `tf.equal`로 `assignments`의 각 원소 위치를 `True`로 표시하는 boolean tensor (Dimension 2000)을 생성
  - `tf.where`로 `True`로 표시된 위치를 값으로 가지는 (2000x1)을 생성
  - `tf.reshape`로 c 군집에 속한 vectors 의 포인트들의 인덱스로 구성된 (1x2000)을 생성
  - `tf.gather`로 c 군집을 이루는 좌표를 모은 (1x2000x1)을 생성
  - `tf.reduce_mean`으로 c 군집에 혹한 모든 점의 평균 값을 가진 (1x2)를 생성

`assignment_values` 텐서의 결과를 확인

```python
data = {"x" : [], "y" : [], "cluster" : []}

for i in range(len(assignment_values)):
    data["x"].append(vectors_set[i][0])
    data["y"].append(vectors_set[i][1])
    data["cluster"].append(assignment_values[i])
    
df = pd.DataFrame(data)
sns.lmplot("x", "y", data=df, fit_reg=False, size=6, hue="cluster", legend=False)
plt.show()
```
