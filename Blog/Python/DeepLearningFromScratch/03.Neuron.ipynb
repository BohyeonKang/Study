- **Sigmoid function (시그모이드 함수)** : `h(x) = 1 / (1 + exp(-x))`

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([1.0, 2.0, 3.0])
```

```python
sigmoid(x)
```

```python
def showMany(x, funcs):
    for func in funcs:
        plt.plot(x, func(x))
    plt.ylim(-0.1, 1.1)
    plt.show() 
```

- **Softmax function (소프트맥스 함수)** : `Yk = exp(Ak) / Sum(exp(Ai~n))`

```python
def softmax(a):
    exp_a = np.exp(a)
    return exp_a / np.sum(exp_a)
```

```python
asm = np.array([1010,1000,990])
softmax(asm)
```

softmax 는 지수계산을 해서 오버플로우가 쉽게 일어난다. 그래서 입력값중 최대값을 빼준 값으로 계산을 해서 오버플로우를 피해가도록 구현할 수 있다.

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    return exp_a / np.sum(exp_a)
```

```python
softmax(asm)
```

```python
import pickle

def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=True, one_hot_label=False)
    return x_test, t_test

def init_network():
    with open("sample_weight.pkl", "rb") as f:
        network = pickle.load(f)
    return network

def predict(network, x):
    W1,W2,W3 = network['W1'], network['W2'], network['W3']
    b1,b2,b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x,W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1,W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3) + b3
    y = softmax(a3)
    
    return y
```

```python
import time
start = time.time()

x, t = get_data()
network = init_network()
accuracy_cnt = 0

print(len(x))

for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1
        
print("Accuracy:" + str(float(accuracy_cnt) / len(x)))

end = time.time() - start 
print(end)
```

### 배치 처리

```python
import time
start = time.time()

x, t = get_data()
network = init_network()

batch_size = 100
accuracy_cnt = 0

print(len(x))

for i in range(0, len(x), batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis=1)
    accuracy_cnt += np.sum(p == t[i:i+batch_size])
        
print("Accuracy:" + str(float(accuracy_cnt) / len(x)))

end = time.time() - start 
print(end)
```
